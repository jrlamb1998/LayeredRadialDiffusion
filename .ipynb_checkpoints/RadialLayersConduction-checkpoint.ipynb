```julia
using Plots
mutable struct Grid
    R   # radius array
    NR  # number of grid points
    DR  # grid spacing
    D   # diffusivity array
    K   # conductivity array
    T   # temperature array
    j   # boundary index
end
```

```julia
function setup(DP, KP, DM, KM)
    RProbe = 0.00476; # radius of probe in cm
    NR = 101;     # number of grid points
    
    ModelSize = 10;   # model size vs probe radius
    
    # grid spacing
    RModel = RProbe * ModelSize;
    DR = RModel/(NR-1);
    R = LinRange(0, RModel, NR); # set the radius coordinates
    
    
    # D and K
    K = KM * ones(NR);
    #K[1:1+(NR-1)/ModelSize] .= KP;
    K[1:trunc(Int, 1+(NR-1)/ModelSize)] .= KP;
    D = DM * ones(NR);
    D[1:trunc(Int, 1+(NR-1)/ModelSize)] .= DP;
    println(D)
        
    # initial temperature
    T = zeros(NR);         # temperature is 0 in the medium
    T[1:trunc(Int, 1+(NR-1)/ModelSize)] .= 1; # temperature is 1 in the probe
    
    j = trunc(Int, 1+(NR-1)/ModelSize);
    
    return Grid(R, NR, DR, D, K, T, j)
end
```

```julia
function rDiffusion(DM, KM) #diffusivity of medium, conductivity of medium
    # grid of thermal properties & coordinates
    DP = 9e-8; #M^2/s
    KP = 0.185; # W/m-K
    G = setup(DP, KP, DM, KM)  # get grid of soil properties and coordinates

    # time step
    Sigma = 0.2; # timestep stability size
    nT = 100;     # number of time steps
    dt = (Sigma*(G.DR^2))/max(DP, DM)
    println(dt)
   
    un = zeros(G.NR); #initialize a temporary array
    for n in 1:nT   #loop for values of n from 1 to nt
        copy!(un,G.T)  #copy existing values of G.T into un
        
        # r = 0
        G.T[1] = un[1] + 4*dt*DM*(un[2] - un[1])*(1/(G.DR^2));
        
        # general interior
        for i in 2:G.NR-1
            G.T[i] = un[i] + dt * ((1/G.R[i])*(1/G.DR^2) * 
                     G.R[i+1]*G.D[i+1]*(un[i+1] - un[i]) - 
                     G.R[i-1]*G.D[i-1]*(un[i] - un[i-1]))
        end
        
        # matching at boundary
#         j = G.j; # find the index of the boundary
        
#         d2tM = G.T[j-2] * (2*KP + 3*KM) -    # derivative in the probe side
#                G.T[j-1] * (2*KP + 6*KM) +
#                G.T[j+1] * (4*KM) -
#                G.T[j+2] * (KM) / (3*(KP + KM)*G.DR^2);
        
#         d2tP = (G.T[j+2] * (2*KM + 3*KP) -    # derivative in the medium side
#                G.T[j+1] * (2*KM + 6*KP) +
#                G.T[j-1] * (4*KP) -
#                G.T[j-2] * (KP)) / (3*(KP + KM)*G.DR^2);
        
#         changeM = ((G.R[j] + 1)/G.R[j]) * DP * d2tM; # change on each side of the boundary
#         changeP = ((G.R[j] + 1)/G.R[j]) * DM * d2tP;

#         G.T[j] = un[j] + (dt/2)*(changeM + changeP); # average changes and do timestep
        
        # outer boundary of the model - make it same as next to the edge
        G.T[end] = un[end] + dt * ((1/G.R[end-1])*(1/G.DR^2) * 
            G.R[end]*G.D[i+1]*(un[end] - un[end-1]) - 
            G.R[end-2]*G.D[end-2]*(un[end-1] - un[end-2]));

    end
    plot(G.R,G.T);    
end
```
